```python
import random
position = 0
walk = [position]
steps = 1000
for i in xrange(steps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)
```

```python
# Exmaple using numpy
```

```python
nsteps = 1000

draws = np.random.randint(0, 2, size=nsteps)

steps = np.where(draws > 0, 1, -1)

walk = steps.cumsum()

print (walk.min())

print (walk.max())

import matplotlib.pyplot as plt

x = np.arange(1000)

plt.plot(x, walk)
plt.show()

# time to (+/-)10 step away from the start point
print (np.abs(walk) >= 10).argmax() # but, not efficient...
```

```python
# 10 times simulation...
```

```python
nwalks = 10
nsteps = 1000

draws = np.random.randint(0, 2, size=(nwalks, nsteps)) # 0 or 1

steps = np.where(draws > 0, 1, -1)

walks = steps.cumsum(1)

print (walks)

x = np.arange(1000)

for walk in walks:
    plt.plot(x, walk)
plt.show()
```

```python
hits30 = (np.abs(walks) >= 30).any(axis=1)
hits30
```

```python
hits30.sum() # the number of simulation which hit 30 or -30
```

```python
crossing_times = (np.abs(walks[hits30]) >= 30).argmax(1)
crossing_times.mean()
```
